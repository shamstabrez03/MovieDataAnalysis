## 2. Count of movies based on certificates (Total no. of movies in each certificate)

```python
data_cert = data
```

```python
data_cert = pd.DataFrame(data_cert.groupby("certificate").movie.count()).reset_index()
```

```python
data_cert
```

```python
cert_value = data_cert["certificate"]
movie_count = data_cert["movie"]

fig = plt.figure(figsize =(30, 15))
plt.bar(cert_value, movie_count, width = 0.6)
 
plt.xlabel("Certificate")
plt.ylabel("Total count of movies")
plt.title("Total Count of Movies per Certificate")
plt.show()
```
